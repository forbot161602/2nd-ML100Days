## 作業

畫出 16 個 bin 的顏色直方圖, 並嘗試回答每個 channel 在 \[16, 32\] 這個 bin 中有多少個像素。

```python
import cv2
import matplotlib.pyplot as plt
from keras.datasets import cifar10
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
image = X_train[0]
```

## 彩色圖片

```python
plt.imshow(image)
plt.show()
```

```python
chans = cv2.split(image) # 把圖像的 RGB channel 切分出來
hists = [cv2.calcHist([chan], [0], None, [16], [0, 256]) for chan in chans]
colors = ('r', 'g', 'b',)

plt.figure(figsize=(8, 6,))
for hist, color in zip(hists, colors):
    plt.plot(hist, color=color)
plt.xlim([0, 16])
plt.xlabel('Bins')
plt.ylabel('# of Pixels')
plt.title('RGB Histogram')
plt.show()
```

```python
for hist, color in zip(hists, colors):
    print('Color {} has {} pixels in the bin [16, 32).'.format(color.upper(), int(hist[1][0])))
```
