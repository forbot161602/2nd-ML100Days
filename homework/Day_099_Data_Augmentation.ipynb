## 範例說明

* 使用 ImageDataGenerator 無限循環批次訓練圖檔，可在產生圖檔前隨機進行平移、旋轉、亮度調整等操作。

```python
import math
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
```

## 準備資料

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

## 定義函數

```python
def image_generator(x, y, batch_size=32, shuffle=True):
    generator = ImageDataGenerator(rotation_range=20, width_shift_range=0.2, height_shift_range=0.2, horizontal_flip=True)
    return generator.flow(x, y, batch_size=batch_size, shuffle=shuffle)

def plot_fixed_sized_thumbnails(images, ncols=8, size=1):
    nrows = int(math.ceil(len(images) / ncols))
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, sharex=True, sharey=True, figsize=(ncols * size, nrows * size,))
    for i in range(nrows):
        for j in range(ncols):
            axes[i][j].imshow(images[i * ncols + j], cmap='rainbow', vmin=0, vmax=1)
            axes[i][j].set_axis_off()
    plt.show()

def display_first_32_images(x, y):
    images, labels = next(image_generator(x, y, shuffle=False)) # 避免圖像不同無法比較，此範例不隨機洗牌。
    plot_fixed_sized_thumbnails(images.astype('int')) # 預處理過的影像資料為浮點數型態，需先轉成整數型態才能顯示。
    print(labels.reshape(-1, 8))
```

### 無限隨機的 CIFAR-10 Generator，一次產生 32 張圖。

```python
display_first_32_images(X_train, y_train)
```
