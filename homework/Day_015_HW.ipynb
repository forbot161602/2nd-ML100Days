```python
# 載入需要的套件
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = np.random.rand(10, 10) * 2 - 1
plt.figure(figsize=(10, 10,))
"""
Your Code Here
"""
sns.heatmap(matrix, cmap=plt.cm.RdYlBu_r, vmin=-1, vmax=1, annot=True)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.uniform(-1, 1, (nrow, ncol,))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0, 1, 2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seaborn 進階 Heatmap
grid = sns.PairGrid(data=plot_data, size=3, diag_sharey=False, hue='index',
                    vars=[x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
"""
grid.map_upper(plt.scatter, alpha=0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap=plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randn(nrow, ncol)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0, 1, 2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seaborn 進階 Heatmap
grid = sns.PairGrid(data=plot_data, size=3, diag_sharey=False, hue='index',
                    vars=[x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
"""
grid.map_upper(plt.scatter, alpha=0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap=plt.cm.OrRd_r)

plt.show()
```
