```python
import numpy as np
import matplotlib.pylab as plt
```

## Sigmoid

Sigmoid 的函數及其導函數

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

def dsigmoid(x):
    return (sigmoid(x) * (1 - sigmoid(x)))
```

```python
x = plt.linspace(-10, 10, 100) # linespace generate an array from start and stop value with requested number of elements.
plt.figure(figsize=(12, 7,))
plt.plot(x, sigmoid(x), 'b')
plt.plot(x, dsigmoid(x), 'b')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=30)
plt.title('Sigmoid')
plt.grid()
plt.show()
```

## Softmax

```python
def softmax(x):
    return np.exp(x) / sum(np.exp(x))
```

```python
x = plt.linspace(-5, 5, 100)
plt.figure(figsize=(12, 7,))
plt.plot(x, softmax(x), 'r')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.01))
plt.title('Softmax')
plt.grid()
plt.show()
```
