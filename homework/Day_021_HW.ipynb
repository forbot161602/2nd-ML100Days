```python
# 做完特徵工程前的所有準備 (與前範例相同)
import copy
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

data_path = './data/part02/'
df_train = pd.read_csv(data_path + 'titanic_train.csv')
df_test = pd.read_csv(data_path + 'titanic_test.csv')

ids = df_test['PassengerId']
train_Y = df_train['Survived']
df_train = df_train.drop(['PassengerId', 'Survived'], axis=1)
df_test = df_test.drop(['PassengerId'], axis=1)
df = pd.concat([df_train, df_test])
df.head()
```

```python
# 只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features: {num_features}\n')

# 削減文字型欄位, 只剩數值型欄位
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df = df[num_features]
df = df.fillna(0)
df.head()
```

```python
# 顯示 Fare 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('bmh')
plt.figure(figsize=(10, 6,))
sns.distplot(df['Fare'][:train_num])
plt.show()
```

```python
# 計算基礎分數
df_mm = MMEncoder.fit_transform(df.astype(float))
train_X = df_mm[:train_num]
estimator = LogisticRegression(solver='lbfgs')
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業1 
* 試著在鐵達尼的票價 (Fare) 欄位中使用對數去偏 (log1p) , 結果是否更好?

答：是

```python
# 將 Fare 取 log1p 後, 看散佈圖, 並計算分數
df_copied = copy.deepcopy(df)
"""
Your Code Here, transform df_copied['Fare'] with log1p
"""
df_copied['Fare'] = np.log1p(df_copied['Fare'])
plt.figure(figsize=(10, 6,))
sns.distplot(df_copied['Fare'][:train_num])
plt.show()

df_copied = MMEncoder.fit_transform(df_copied.astype(float))
train_X = df_copied[:train_num]
estimator = LogisticRegression(solver='lbfgs')
print(cross_val_score(estimator, train_X, train_Y, cv=5).mean())
```

# 作業2
* 最後的 boxcox 區塊直接執行會造成錯誤, 起因為輸入值有負值, 請問如何修正後可以使用 boxcox? (Hint: 試圖修正資料)

```python
# 將 Fare 取 boxcox 後, 看散佈圖, 並計算分數 (執行會有 error, 請試圖修正)
from scipy import stats
df_copied = copy.deepcopy(df)
"""
Your Code Here, fix the error
"""
df_copied.loc[df_copied['Fare'] <= 0, 'Fare'] = df_copied['Fare'].median()
df_copied['Fare'] = stats.boxcox(df_copied['Fare'])[0]
plt.figure(figsize=(10, 6,))
sns.distplot(df_copied['Fare'][:train_num])
plt.show()

df_copied = MMEncoder.fit_transform(df_copied.astype(float))
train_X = df_copied[:train_num]
estimator = LogisticRegression(solver='lbfgs')
print(cross_val_score(estimator, train_X, train_Y, cv=5).mean())
```
