```python
import numpy as np
import matplotlib.pylab as plt
```

## ReLU

```python
def ReLU(x):
    return (x > 0) * abs(x)

def dReLU(x):
    return (x > 0) * 1
```

```python
x = plt.linspace(-5, 5, 10000)
plt.figure(figsize=(12, 7,))
plt.plot(x, ReLU(x), 'b')
plt.plot(x, dReLU(x), 'b')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.text(-0.5, 4, r'$f(x)=max(0, x)$', fontsize=30)
plt.title('ReLU')
plt.grid()
plt.show()
```
