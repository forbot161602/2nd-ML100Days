```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
app_train.sample(100)
```

```python
cnt_fam_members = app_train['CNT_FAM_MEMBERS']
cnt_fam_members.describe()
```

```python
print('The followings are statistics about data of the column {}:'.format(cnt_fam_members.name))
print('count:  {:10d}'.format(int(cnt_fam_members.count())))
print('sum:    {:10d}'.format(int(cnt_fam_members.sum())))
print('mean:   {:10d}'.format(int(cnt_fam_members.mean())))
print('median: {:10d}'.format(int(cnt_fam_members.median())))
print('mode:   {:10d}'.format(int(cnt_fam_members.mode())))
print('max:    {:10d}'.format(int(cnt_fam_members.max())))
print('Q1:     {:10d}'.format(int(cnt_fam_members.quantile(0.25))))
print('Q2:     {:10d}'.format(int(cnt_fam_members.quantile(0.50))))
print('Q3:     {:10d}'.format(int(cnt_fam_members.quantile(0.75))))
print('min:    {:10d}'.format(int(cnt_fam_members.min())))
print('std:    {:10.2f}'.format(cnt_fam_members.std()))
print('var:    {:10.2f}'.format(cnt_fam_members.var()))
print('skew:   {:10.2f}'.format(cnt_fam_members.skew()))
print('kurt:   {:10.2f}'.format(cnt_fam_members.kurt()))
```

```python
print('The correlation coefficient between the column CNT_FAM_MEMBERS and CNT_CHILDREN is '
      '{:1.2f}.'.format(cnt_fam_members.corr(app_train['CNT_CHILDREN'])))
```
